```python
# Integer number
num = 100
print(num)
print("Data type of variable num is", type(num))

# float number
fnum = 34.45
print(fnum)
print("Data type of variable fnum is", type(fnum))

# complex number
cnum = 3 + 4j
print(cnum)
print("Data type of variable cnum is", type (cnum))
```

```python
# Simple Interest

P = float(input("Principal: "))
R = float(input("Rate (%): "))
T = float(input("Time in years: "))

A = P * (1 + (R/100) * T)

print("SI = ", A)
```

```python
# Compound Interest

P = float(input("Principal: "))
R = float(input("Rate in decimal: "))
n = int(input("Number: "))
t = float(input("Time in years: "))

A = P * (1 + R/n) ** (n*t)

print("CI =", A)
```

```python
# Annuity Plan

PMT = float(input("PMT: "))
R = float(input("Rate in decimal: "))
n = int(input("Number: "))
t = float(input("Time: "))

A = PMT * (((1 + R/n) ** (n*t) - 1) / (R/n))

print("Annuity =", A)
```
