```python
def printinfo(name, age):
    print("Name: ", name)
    print("Age: ", age)
    return

printinfo(age=50, name="miki")
```

```python

def printinfo(name, age=35):
    print("Name: ", name)
    print("Age: ", age)
    return

printinfo(age=50, name="miki")
printinfo(name="miki")
```

```python
def printinfo(arg1, *vartuple):
    print("Output is: ")
    print(arg1)
    print(*vartuple)
    for var in vartuple:
        print(var)
    return

printinfo(10)
printinfo(70, 60, 50, 4.0)
```
