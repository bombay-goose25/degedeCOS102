```python

import tkinter as tk
from tkinter import messagebox

def welcomeMessage(username):
    """Display welcome window after successful login"""
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x200")
    
    label_1 = tk.Label(window, text=f"Welcome {username}", font=("Arial", 14))
    label_1.pack(pady=20)
    
    label_2 = tk.Label(window, text="This is Python GUI with Tkinter")
    label_2.pack()

    logout_btn = tk.Button(window, text="Logout", command=window.destroy)
    logout_btn.pack(pady=20)

def submit():
    """Handle login form submission"""
    username = username_entry.get()
    password = password_entry.get()
    
    if username == "Mary" and password == "cos102":
        welcomeMessage(username)
    else:
        messagebox.showerror("Login Error", "Invalid username or password")

def clear_form():
    """Clear the entry fields"""
    username_entry.delete(0, tk.END)
    password_entry.delete(0, tk.END)

root = tk.Tk()
root.title("Login Form")
root.geometry("500x250")
root.configure(bg="lightblue")

username_label = tk.Label(root, text="Username:", bg="lightblue", font=("Arial", 12))
username_label.pack(pady=(20, 5))
username_entry = tk.Entry(root, width=30)
username_entry.pack()

password_label = tk.Label(root, text="Password:", bg="lightblue", font=("Arial", 12))
password_label.pack(pady=(10, 5))
password_entry = tk.Entry(root, show="*", width=30)
password_entry.pack()

submit_button = tk.Button(root, text="Submit", command=submit, bg="green", fg="white", width=15)
submit_button.pack(pady=10)

clear_button = tk.Button(root, text="Clear", command=clear_form, bg="orange", width=15)
clear_button.pack()

root.mainloop()
```
