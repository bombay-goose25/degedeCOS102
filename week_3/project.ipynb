```python
import tkinter as tk
from tkinter import messagebox
import csv

employees = []

def load_data():
    try:
        with open('GIG-logistics.csv', 'r') as file:
            reader = csv.reader(file)
            header = next(reader) 
            for row in reader:
                employees.append(row)
        return True
    except FileNotFoundError:
        messagebox.showerror("Error", "Dataset file 'GIG-logistics.csv' not found!")
        return False

def get_dept_members(dept):
    members = []
    for emp in employees:
        if emp[2] == dept: 
            members.append(emp[1]) 
    return members

def display_dept_members(dept):
    members_text.delete(1.0, tk.END)
    members = get_dept_members(dept)
    members_text.insert(tk.END, f"Department: {dept}\n")
    members_text.insert(tk.END, "-" * 30 + "\n\n")
    
    for member in members:
        members_text.insert(tk.END, f"• {member}\n")
def verify_employee():
    name = name_entry.get()
    dept = dept_var.get()

    if name == "" or dept == "":
        messagebox.showwarning("Input Error", "Please enter name and select department!")
        return

    employee_exists = False
    employee_info = ""
    
    for emp in employees:
        if emp[1].lower() == name.lower() and emp[2] == dept:
            employee_exists = True
            employee_info = f"Name: {emp[1]}\nDepartment: {emp[2]}\nPosition: {emp[3]}"
            break

    if employee_exists:
        messagebox.showinfo("Welcome!", f"Welcome {name}!\n\n{employee_info}")

        status_label.config(text="✓ EMPLOYEE VERIFIED", fg="green")

        display_dept_members(dept)
    else:
        messagebox.showinfo("Not Found", f"Sorry, {name} is not an employee in the {dept} department.")
        status_label.config(text="✗ EMPLOYEE NOT FOUND", fg="red")
        members_text.delete(1.0, tk.END)
        members_text.insert(tk.END, "No employee found with these details.")

def clear_form():
    name_entry.delete(0, tk.END)
    dept_var.set("")
    status_label.config(text="Ready", fg="gray")
    members_text.delete(1.0, tk.END)

def main():
    global name_entry, dept_var, status_label, members_text, root
    if not load_data():
        return
    root = tk.Tk()
    root.title("GIG Logistics - Employee Verification")
    root.geometry("550x450")
    title = tk.Label(root, text="GIG LOGISTICS", font=("Arial", 18, "bold"))
    title.pack(pady=10)
    
    subtitle = tk.Label(root, text="Employee Verification System", font=("Arial", 10))
    subtitle.pack()
    input_frame = tk.Frame(root)
    input_frame.pack(pady=20)
    name_label = tk.Label(input_frame, text="Name:")
    name_label.grid(row=0, column=0, padx=5, pady=5, sticky="e")
    
    name_entry = tk.Entry(input_frame, width=30)
    name_entry.grid(row=0, column=1, padx=5, pady=5)
    dept_label = tk.Label(input_frame, text="Department:")
    dept_label.grid(row=1, column=0, padx=5, pady=5, sticky="e")
    dept_list = []
    
    for emp in employees:
        if emp[2] not in dept_list:
            dept_list.append(emp[2])
    
    dept_var = tk.StringVar()
    dept_dropdown = tk.OptionMenu(input_frame, dept_var, *dept_list)
    dept_dropdown.config(width=27)
    dept_dropdown.grid(row=1, column=1, padx=5, pady=5)
    button_frame = tk.Frame(root)
    button_frame.pack(pady=10)
    verify_btn = tk.Button(button_frame, text="VERIFY", command=verify_employee, bg="green", fg="white", width=15)
    verify_btn.pack(side="left", padx=10)
    clear_btn = tk.Button(button_frame, text="CLEAR", command=clear_form, bg="red", fg="white", width=15)
    clear_btn.pack(side="left", padx=10)
    status_label = tk.Label(root, text="Ready", fg="gray", font=("Arial", 10))
    status_label.pack(pady=10)
    members_label = tk.Label(root, text="Department Members:", font=("Arial", 10, "bold"))
    members_label.pack(anchor="w", padx=20)
    members_text = tk.Text(root, height=10, width=60, wrap="word")
    members_text.pack(pady=5, padx=20)
    root.mainloop()
if __name__ == "__main__":
    main()
```
