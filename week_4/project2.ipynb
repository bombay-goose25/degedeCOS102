```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("titanic.csv")
print(df.head())
```

```python
df = df.dropna(subset=['Age', 'Fare'])
df = df[df['Fare'] > 0]
df = df[df['Fare'] < 500]
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
plt.figure()
sns.histplot(df['Age'], bins=30)

plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
print("The results show that most passengers were concentrated in the young adult age and fewer passengers at very young or older ages.")
print("The distribution is slightly right-skewed showing that there are fewer older passengers than to younger ones.")
```

```python
plt.figure()
sns.boxplot(x=df['Fare'])

plt.title("Boxplot of Ticket Fare")

plt.show()
```

```python
print("The plot shows that the median fare is low showing that most passengers paid lower ticket prices.")
print("There is a wide spread in fare values with several high value outliers above the upper whisker.")
```

```python
plt.figure()
sns.scatterplot(x='Age', y='Fare', data=df)

plt.title("Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

```python
print("The result show a weak relationship between fare and age as the data points are widely dispersed.")
```
