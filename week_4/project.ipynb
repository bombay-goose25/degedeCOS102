```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("AB_NYC_2019.csv")

print(df.head())
```

```python
df = df.dropna(subset=['price', 'number_of_reviews', 'availability_365'])
```

```python
df = df[df['price'] < 1000]
```

```python
plt.figure()
plt.hist(df['price'], bins=50)

plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

```python
print("The result show that majority fall within the lower price ranges with fewer at higher prices.")
print("The distribution is positively skewed indicating the presence of a small number of outliers.")
```

```python
plt.figure()
sns.boxplot(x=df['price'])

plt.title("Boxplot of Price")

plt.show()
```

```python
print("The plot shows that the median price is within the lower range and that most listings have affordable prices.")
```

```python
plt.figure()
plt.scatter(df['number_of_reviews'], df['price'])

plt.title("Price vs Number of Reviews")
plt.xlabel("Number of Reviews")
plt.ylabel("Price")

plt.show()
```

```python
print("The result show a weak correlation between these price and number of reviews because the data points are packed")
```
